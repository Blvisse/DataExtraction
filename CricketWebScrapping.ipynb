```
import pandas as pd
import numpy as np
import re
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
import csv
```

We will use data from howstat.com. cricket player statistics to be precise.
We get the url and store it in the url parameter we use the request Libraray to make the request to the url. we use soup to parse the html and then search for the particular identifier we wish to cillect data from in this case it's a table with class TableLined


---

We are only going to get the players whose name starts with A

```
url='http://howstat.com/cricket/Statistics/Players/PlayerList.asp?Country=ALL&Group=A'
pageHtml=uReq(url)
soup=soup(pageHtml,'html.parser')
table=soup.find('table',{'class','TableLined'})
```

We are accessing the individual cells by passing a loop through them.
The first loop ientifies rows in the table(tr),we then find the data in the rows whihc we find by the attribute(td)then append it to the data list 

```
for row in table:
  rows=table.find_all('tr')
for tr in rows:
  data=[]
  cols=tr.find_all('td')
  for td in cols:
    data.append(td.text.strip())
```

We proceed to save the data in a csv file for future access

```
with open('Cricket.csv','a',newline='')as csvfile:
  f=csv.writer(csvfile)

  for row in table:
    rows=table.find_all('tr')   
  for tr in rows:
    data=[]
    cols=tr.find_all('td')
    for td in cols:
      data.append(td.text.strip())
    f.writerow(data)
    print(data)
```
